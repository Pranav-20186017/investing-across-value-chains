```python
def add_stock_returns(df, stock_list):
    # Dictionary to hold our new return columns before we add them to the DataFrame
    returns_dict = {}
    
    for stock in stock_list:
        for price_type in ['Open', 'High', 'Low', 'Close']:
            col_name = f"{stock}_{price_type}"
            if col_name in df.columns:
                return_col_name = f"{stock}_Returns_{price_type[:2].upper()}"
                # Instead of adding the column directly to df, we add it to our dictionary
                returns_dict[return_col_name] = df[col_name].pct_change()

    # Now we concatenate all our new columns to the DataFrame at once
    returns_df = pd.DataFrame(returns_dict)
    df = pd.concat([df, returns_df], axis=1)
    
    return df
```

```python
def analyze_portfolio(merged_df, price_col, risk_free_rate=0.0738):
    # Extracting the specified price columns and summing them to get the portfolio value
    price_cols = [col for col in merged_df.columns if price_col in col]
    portfolio_value = merged_df[price_cols].sum(axis=1)

    # Plotting the portfolio value over time with enhancements
    plt.figure(figsize=(12, 6))
    portfolio_value.plot(title=f"Portfolio Value Over Time - {price_col}", legend=True, color='blue', linewidth=2)
    plt.xlabel("Date")
    plt.ylabel("Portfolio Value")
    plt.grid(True)
    plt.legend([f"Portfolio Value - {price_col}"])
    plt.show()

    # Assuming each year has approximately 252 trading days
    trading_days = 252

    # Calculating Yearly ROI
    yearly_roi = portfolio_value.pct_change(trading_days).dropna()

    # Calculating CAGR
    total_days = (portfolio_value.index[-1] - portfolio_value.index[0]).days
    cagr = (portfolio_value.iloc[-1] / portfolio_value.iloc[0]) ** (365.0 / total_days) - 1

    # Calculating Total ROI
    total_roi = (portfolio_value.iloc[-1] - portfolio_value.iloc[0]) / portfolio_value.iloc[0]

    # Calculating daily returns
    daily_returns = portfolio_value.pct_change().dropna()

    # Calculating Sharpe Ratio
    sharpe_ratio = (daily_returns.mean() - risk_free_rate / trading_days) / daily_returns.std() * np.sqrt(trading_days)

    print(f"Yearly ROI for {price_col}:\n{yearly_roi}")
    print(f"CAGR for {price_col}: {cagr:.2%}")
    print(f"Total ROI for {price_col}: {total_roi:.2%}")
    print(f"Sharpe Ratio for {price_col}: {sharpe_ratio:.2f}")
```

```python
analyze_portfolio(merged_df, 'Open')
```
